```python
from ggplot import mtcars;
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(mtcars["wt"], mtcars["mpg"], label = "data", color = "red", marker = "o")
sns.regplot(x = "wt", y = "mpg", data = mtcars, label = "order 1", color = "blue", order =1, scatter = None)
sns.regplot(x = "wt", y = "mpg", data = mtcars, label = "order 2", color = "green", order = 2, scatter = None)
plt.legend(loc = "upper right")
plt.show()
```

```python
sns.lmplot(x = "wt", y = "mpg", data = mtcars, hue = "name", palette = "Set1")
plt.show()
```

```python
# convert "cyl" column data type "int" to "str" with using "apply()"
mtcars["cyl"] = mtcars["cyl"].apply(str)
mtcars.info()
```

```python
# convert data variable name "mtcars" to "auto"
auto = mtcars
auto.info()
```

```python
plt.subplot(2, 1, 1)
sns.stripplot(x = "cyl", y = "hp", data = auto)
plt.subplot(2, 1, 2)
sns.stripplot(x = "cyl", y = "hp", data = auto, jitter = True, size = 3)
plt.show()
```

```python
auto["gear"] = auto["gear"].apply(str)
auto["am"] = auto["am"].apply(str)
auto.info()
```

```python
plt.subplot(2, 1, 1)
sns.violinplot(x = "cyl", y = "hp", data = auto)
plt.subplot(2, 1, 2)
sns.violinplot(x = "cyl", y = "hp", data = auto, inner = None, color= "lightgray")
plt.subplot(2, 1, 2)
sns.stripplot(x = "cyl", y = "hp", data = auto, jitter = True, size = 1.5)
plt.show()
```

```python
# generate a joint plot of "hp" and "mpg"
sns.jointplot(x = "hp", y = "mpg", data = auto)
plt.show()
```

```python
# generate a joint plot of "hp" and "mpg" using a hexbin plot
sns.jointplot(x = "hp", y = "mpg", data = auto, kind = "hex")
plt.show()
```
